# Bar Charts

```python
google = data.DataReader(name='GOOG', data_source='yahoo', start='2004-01-01', end='2011-01-01')
google.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 300:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google['Close'].apply(rank_performance).value_counts()
```

```python
google['Close'].apply(rank_performance).value_counts().plot()
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind='line')
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind='bar')
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind='barh')
```

# Pie Charts

```python
apple = data.DataReader(name='AAPL', data_source='yahoo', start='2012-01-01', end='2016-01-01')
apple.head(3)
```

```python
apple['Close'].mean()
```

```python
def rank_performacne(stock_price):
    if stock_price >= 90.54:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
apple['Close'].apply(rank_performacne).value_counts().plot(kind='pie', legend=True)
```

# Histograms

```python
google = data.DataReader(name='GOOG', data_source='yahoo', start='2012-01-01', end='2016-01-01')
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
custom_round(490)
```

```python
google['High'].apply(custom_round).value_counts().sort_index()
```

```python
google['High'].apply(custom_round).nunique()
```

```python

google['High'].apply(custom_round).plot(kind='hist')
```

```python
n = google['High'].apply(custom_round).nunique()
google['High'].apply(custom_round).plot(kind='hist', bins=n)
```
